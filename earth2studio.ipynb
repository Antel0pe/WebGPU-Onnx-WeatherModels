```python
from earth2studio.models.px import Pangu3
from earth2studio.data import GFS
from earth2studio.io import NetCDF4Backend
from earth2studio.run import deterministic as run

# Load the pretrained Pangu3 model package and construct the model
model = Pangu3.load_model(Pangu3.load_default_package())

# Data source – lets Earth2Studio fetch & prep inputs for you
ds = GFS()  # public, no credentials needed

# Output backend – write forecast to a NetCDF file
io = NetCDF4Backend("output.nc")

# Run a deterministic forecast:
# - start_time: any date/time that exists in GFS (string or datetime-like)
# - steps: number of model steps to take into the future
#   Pangu3 is a 3-hour lead model, so:
#   total forecast lead = steps * 3 hours
start_times = ["2024-01-01T00:00"]  # or any valid time the data source supports
steps = 40                          # 40 * 3h = 120h = 5 days

run(start_times, steps, model, ds, io)
```
